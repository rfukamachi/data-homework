```python
%matplotlib notebook
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_raw = pd.read_csv(mouse_drug_data_to_load)
mouse_drug_raw.head()

ctrial_raw = pd.read_csv(clinical_trial_data_to_load)
ctrial_raw.head()

# Combine the data into a single dataset
combo_raw = mouse_drug_raw.merge(ctrial_raw, on = "Mouse ID", how = "left")

# Display the data table for preview
combo_raw.head()

```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
DrugTimeGrp_df = pd.DataFrame({
    "Avg Tumor Vol (mm3)": combo_raw.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]
}).reset_index()     #reset_index removes the group by index

# Preview DataFrame
DrugTimeGrp_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame

SemDrugTimeGrp_df = pd.DataFrame({
    "SEM Tumor Vol (mm3)" : combo_raw.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]
}).reset_index()

SemDrugTimeGrp_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
AvgTumorVolPivot = DrugTimeGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'Avg Tumor Vol (mm3)')

# Preview that Reformatting worked
AvgTumorVolPivot
```

```python
# Pivot the SEM table:
SEMDrugTimePivot = SemDrugTimeGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'SEM Tumor Vol (mm3)')
SEMDrugTimePivot

# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

# Define x axis:
x_axis = np.arange(0, max(combo_raw["Timepoint"])+1, 5)
x_axis

# THIS WORKS BUT THE FMT WONT CHANGE BTW DRUGS:
# # DrugList = combo_raw["Drug"].unique()     //commented out because we only needed 4 mentioned in the README
# DrugList = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']

# for x in DrugList:
#     means = AvgTumorVolPivot[x]
#     standard_errors = SEMDrugTimePivot[x]
#     ax.errorbar(x_axis, means, standard_errors, fmt = 'd')
#     ax.plot(x_axis, AvgTumorVolPivot[x], label = x)



DrugList = pd.DataFrame({
    'Drug': ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'],
    'fmt': ['o', '^', 's', 'd'],
    'color': ['red', 'blue', 'green', 'black']
})
# DrugList

for index, row in DrugList.iterrows():
    means = AvgTumorVolPivot[row['Drug']]
    standard_errors = SEMDrugTimePivot[row['Drug']]
    ax.errorbar(x_axis, means, standard_errors, fmt = row['fmt'], color = row['color'])
    ax.plot(x_axis, AvgTumorVolPivot[row['Drug']], row['fmt'] + ':', color = row['color'], label = row['Drug'])
    

    
ax.set_xlabel("Timepoint (Days)")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_title("Tumor Response to Treatment")
plt.legend(loc = "upper left", fancybox = True)
plt.grid()


# Save the Figure
TumorRespToTxChart = fig

# Show the Figure
TumorRespToTxChart.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
DrugMetsGrp_df = pd.DataFrame({
    "Avg Mets Sites": combo_raw.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]
}).reset_index()

# Preview DataFrame
# DrugMetsGrp_df.head()
Display_DrugMets = DrugMetsGrp_df.groupby(["Drug", "Timepoint"])
Display_DrugMets.mean().head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
SemDrugMetsGrp_df = pd.DataFrame({
    "SEM Mets Sites" : combo_raw.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]
}).reset_index()

# Preview DataFrame
Display_SEMDrugMets = SemDrugMetsGrp_df.groupby(["Drug", "Timepoint"])
Display_SEMDrugMets.mean().head()
```

```python
# # Minor Data Munging to Re-Format the Data Frames
AvgMetsSitesPivot = DrugMetsGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'Avg Mets Sites')

# Preview that Reformatting worked
AvgMetsSitesPivot
```

```python
# Pivot the SEM table:
SEMDrugMetsPivot = SemDrugMetsGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'SEM Mets Sites')
SEMDrugMetsPivot

# Generate the Plot (with Error Bars)
fig2, ax2 = plt.subplots()

# Define x axis:
x_axis2 = np.arange(0, max(combo_raw["Timepoint"])+1, 5)
x_axis2


DrugList = pd.DataFrame({
    'Drug': ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'],
    'fmt': ['o', '^', 's', 'd'],
    'color': ['red', 'blue', 'green', 'black']
})
# DrugList

for index, row in DrugList.iterrows():
    means2 = AvgMetsSitesPivot[row['Drug']]
    standard_errors2 = SEMDrugMetsPivot[row['Drug']]
    ax2.errorbar(x_axis2, means2, standard_errors2, fmt = row['fmt'], color = row['color'])
    ax2.plot(x_axis2, AvgMetsSitesPivot[row['Drug']], row['fmt'] + ':', color = row['color'], label = row['Drug'])
    

    
ax2.set_xlabel("Treatment Duration (Days)")
ax2.set_ylabel("Metastatic Sites")
ax2.set_title("Metastatic Spread During Treatment")
plt.legend(loc = "upper left", fancybox = True)
plt.grid()


# Save the Figure
MetSitesToTxChart = fig

# Show the Figure
MetSitesToTxChart.show()
```
