```python
%matplotlib notebook
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_raw = pd.read_csv(mouse_drug_data_to_load)
mouse_drug_raw.head()

ctrial_raw = pd.read_csv(clinical_trial_data_to_load)
ctrial_raw.head()

# Combine the data into a single dataset
combo_raw = mouse_drug_raw.merge(ctrial_raw, on = "Mouse ID", how = "left")

# Display the data table for preview
combo_raw.head()

```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
DrugTimeGrp_df = pd.DataFrame({
    "Avg Tumor Vol (mm3)": combo_raw.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]
}).reset_index()     #reset_index removes the group by index

# Preview DataFrame
DrugTimeGrp_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame

SemDrugTimeGrp_df = pd.DataFrame({
    "SEM Tumor Vol (mm3)" : combo_raw.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]
}).reset_index()

SemDrugTimeGrp_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
AvgTumorVolPivot = DrugTimeGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'Avg Tumor Vol (mm3)')

# Preview that Reformatting worked
AvgTumorVolPivot
```

```python
# Pivot the SEM table:
SEMDrugTimePivot = SemDrugTimeGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'SEM Tumor Vol (mm3)')
SEMDrugTimePivot

# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

# Define x axis:
x_axis = np.arange(0, max(combo_raw["Timepoint"])+1, 5)
x_axis

# THIS WORKS BUT THE FMT WONT CHANGE BTW DRUGS:
# # DrugList = combo_raw["Drug"].unique()     //commented out because we only needed 4 mentioned in the README
# DrugList = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']

# for x in DrugList:
#     means = AvgTumorVolPivot[x]
#     standard_errors = SEMDrugTimePivot[x]
#     ax.errorbar(x_axis, means, standard_errors, fmt = 'd')
#     ax.plot(x_axis, AvgTumorVolPivot[x], label = x)



DrugList = pd.DataFrame({
    'Drug': ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'],
    'fmt': ['o', '^', 's', 'd'],
    'color': ['red', 'blue', 'green', 'black']
})
# DrugList

for index, row in DrugList.iterrows():
    means = AvgTumorVolPivot[row['Drug']]
    standard_errors = SEMDrugTimePivot[row['Drug']]
    ax.errorbar(x_axis, means, standard_errors, fmt = row['fmt'], color = row['color'])
    ax.plot(x_axis, AvgTumorVolPivot[row['Drug']], row['fmt'] + ':', color = row['color'], label = row['Drug'])
    

    
ax.set_xlabel("Timepoint (Days)")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_title("Tumor Response to Treatment")
plt.legend(loc = "upper left", fancybox = True)
plt.grid()


# Save the Figure
TumorRespToTxChart = fig

# Show the Figure
TumorRespToTxChart.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
DrugMetsGrp_df = pd.DataFrame({
    "Avg Mets Sites": combo_raw.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]
}).reset_index()

# Preview DataFrame
# DrugMetsGrp_df.head()
Display_DrugMets = DrugMetsGrp_df.groupby(["Drug", "Timepoint"])
Display_DrugMets.mean().head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
SemDrugMetsGrp_df = pd.DataFrame({
    "SEM Mets Sites" : combo_raw.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]
}).reset_index()

# Preview DataFrame
Display_SEMDrugMets = SemDrugMetsGrp_df.groupby(["Drug", "Timepoint"])
Display_SEMDrugMets.mean().head()
```

```python
# # Minor Data Munging to Re-Format the Data Frames
AvgMetsSitesPivot = DrugMetsGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'Avg Mets Sites')

# Preview that Reformatting worked
AvgMetsSitesPivot
```

```python
# Pivot the SEM table:
SEMDrugMetsPivot = SemDrugMetsGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'SEM Mets Sites')
SEMDrugMetsPivot

# Generate the Plot (with Error Bars)
fig2, ax2 = plt.subplots()

# Define x axis:
x_axis2 = np.arange(0, max(combo_raw["Timepoint"])+1, 5)
x_axis2


DrugList = pd.DataFrame({
    'Drug': ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'],
    'fmt': ['o', '^', 's', 'd'],
    'color': ['red', 'blue', 'green', 'black']
})
# DrugList

for index, row in DrugList.iterrows():
    means2 = AvgMetsSitesPivot[row['Drug']]
    standard_errors2 = SEMDrugMetsPivot[row['Drug']]
    ax2.errorbar(x_axis2, means2, standard_errors2, fmt = row['fmt'], color = row['color'])
    ax2.plot(x_axis2, AvgMetsSitesPivot[row['Drug']], row['fmt'] + ':', color = row['color'], label = row['Drug'])
    

    
ax2.set_xlabel("Treatment Duration (Days)")
ax2.set_ylabel("Metastatic Sites")
ax2.set_title("Metastatic Spread During Treatment")
plt.legend(loc = "upper left", fancybox = True)
plt.grid()


# Save the Figure
MetSitesToTxChart = fig

# Show the Figure
MetSitesToTxChart.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
MiceCountGrp_df = pd.DataFrame({
    "Mouse Count": combo_raw.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]
}).reset_index()     #reset_index removes the group by index

# Preview DataFrame
MiceCountGrp_df.head()
```

```python

MiceCountPivot = MiceCountGrp_df.pivot(index = 'Timepoint' , columns = 'Drug', values = 'Mouse Count')
MiceCountPivot
```

```python
# Generate the Plot (Accounting for percentages)
fig3, ax3 = plt.subplots()

# Define x axis:
x_axis3 = np.arange(0, max(combo_raw["Timepoint"])+1, 5)
x_axis3


DrugList = pd.DataFrame({
    'Drug': ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'],
    'fmt': ['o', '^', 's', 'd'],
    'color': ['red', 'blue', 'green', 'black']
})
# DrugList

for index, row in DrugList.iterrows():
    mousepct = round((MiceCountPivot[row['Drug']]/25)*100)
    ax3.plot(x_axis3, mousepct, row['fmt'] + ':', color = row['color'], label = row['Drug'])
        
ax3.set_xlabel("Time (Days)")
ax3.set_ylabel("Survival Rate (%)")
ax3.set_title("Survival During Treatment")
plt.legend(loc = "bottom left", fancybox = True)
plt.grid()


# Save the Figure
SurvivalRateChart = fig

# Show the Figure
SurvivalRateChart.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

#Retrieve complete drug list:
CompleteDrugList = combo_raw["Drug"].unique()  

#Sort it:
CompleteDrugList.sort()
CompleteDrugList

Pchange = []
for x in CompleteDrugList:
    Pchange.append(round(((AvgTumorVolPivot[x][45] - AvgTumorVolPivot[x][0])/AvgTumorVolPivot[x][0])*100, 6))

Pchange

Pchange_df = pd.DataFrame({
    'Drug': CompleteDrugList,
    'PChange': Pchange
})


# Display the data to confirm
# Pchange_df = Pchange_df.set_index("Drug")
Pchange_df
```

```python
# Store all Relevant Percent Changes into a Tuple
Drugs = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']

IndexedPchange_df = Pchange_df.set_index("Drug")
pchange_tp = IndexedPchange_df.loc[Drugs, "PChange"]
pchange_tp 


# Splice the data between passing and failing drugs

#PASSES:
PassedDrugs = ['Capomulin']
Passed_tp = IndexedPchange_df.loc[PassedDrugs, "PChange"]
Passed_tp


#FAILS:
FailedDrugs = ['Infubinol', 'Ketapril', 'Placebo']
Failed_tp = IndexedPchange_df.loc[FailedDrugs, "PChange"]
Failed_tp



#PLOT THE PASSES AND FAILS:
fig4, ax4 = plt.subplots()
PassedBars = ax4.bar(PassedDrugs, Passed_tp, color = 'green', align = 'center', width = 1)
FailedBars = ax4.bar(FailedDrugs, Failed_tp, color = 'red', align = 'center', width = 1)



# # Orient widths. Add labels, tick marks, etc. 
ax4.set_ylabel('% Tumor Volume Change')
ax4.set_title("Tumor Change Over 45 Day Treatment")
plt.xticks(np.arange(4) + 0.5, Drugs)
plt.yticks(np.arange(-20, 61, step = 20))
plt.xlim(-0.75, 3.75)
plt.ylim(-30, 70)
plt.grid()

# Use functions to label the percentages of changes
def autolabel(rects):
    
    for rect in rects:
        height = rect.get_height()
        
        if rect.get_height() < 0:
            ax4.text(rect.get_x() + rect.get_width()/2., -5, str(int(height)) + '%', ha='center', va='bottom', color = 'white')
        else:          
            ax4.text(rect.get_x() + rect.get_width()/2., 5, str(int(height)) + '%', ha='center', va='bottom', color = 'white')

            
# Call functions to implement the function calls
autolabel(PassedBars)
autolabel(FailedBars)


# Save the Figure
BarChart = fig

# Show the Figure
BarChart.show()
```
