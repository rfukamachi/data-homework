```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Total number of players:

ttl_plyrs = purchase_data["SN"].nunique()
ttl_plyrs
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#########################################################################
#Total number of unique item IDs:
ttl_items = purchase_data["Item ID"].nunique()
# ttl_items

#Average Price:
items_df = purchase_data[["Item ID", "Price"]]
avgprice = items_df["Price"].mean()
avgprice = "${:.2f}".format(avgprice)
# avgprice

#Number of Purchases:
ttl_purchs = purchase_data["Purchase ID"].nunique()
# ttl_purchs

#Total Revenues:
ttl_rev = purchase_data["Price"].sum()
# ttl_rev

#########################################################################
#Summary Table: 
SummaryItems = pd.DataFrame({
    "Number of Unique Items": [ttl_items],
    "Average Price": [avgprice],
    "Number of Purchases": [ttl_purchs],
    "Total Revenue": [ttl_rev]
})

SummaryItems
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
##########################################################
#Gender Table:
UserTable = purchase_data.drop_duplicates(subset = "SN", keep = "first")
GenderCount = UserTable["Gender"].value_counts()
GenderCount.head()

#Gender Percentages:
GenderPercent = (GenderCount/ttl_plyrs * 100).round(2)
GenderPercent

#Summary Table:
SummaryGender = pd.DataFrame({
    "Total Count": GenderCount,
    "Percentage of Players": GenderPercent
})

SummaryGender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#########################################################################
# Purchase Counts by Gender:
#########################################################################
#Purchase/Gender Table:
Purch_df = purchase_data[["Gender", "Price"]]

#Purchase Count by Gender:
PurchCountByGender = Purch_df["Gender"].value_counts()
PurchCountByGender

#########################################################################
#Average Purchase Price by Gender:
#########################################################################
#Total Purchase Price by gender:
TtlPurchPriceByGender = Purch_df.groupby(["Gender"]).sum()["Price"]
TtlPurchPriceByGender

#Average Purchase Price by Gender:
AvgPurchPriceByGender = (TtlPurchPriceByGender/PurchCountByGender).map("${:.2f}".format)
AvgPurchPriceByGender

#########################################################################
#Average Total Purchase per Person by Gender:
#########################################################################
# Purchase count per SN:
GenderCount
#Total Purchase Prie by Gender;
TtlPurchPriceByGender

AvgPurchPricePerSN_ByGender = (TtlPurchPriceByGender/GenderCount).map("${:.2f}".format)
AvgPurchPricePerSN_ByGender


#########################################################################
#Summary Table:
#########################################################################
SummaryPurchByGender = pd.DataFrame({
    "Purchase Count": PurchCountByGender,
    "Average Purchase Price": AvgPurchPriceByGender,
    "Total Purchase Value": TtlPurchPriceByGender.map("${:.2f}".format),
    "Avg Total Purchase per Person": AvgPurchPricePerSN_ByGender
})

SummaryPurchByGender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#########################################################################
#Age Demographics:
#########################################################################
#Get Max Age to get the last bin:
maxAge = purchase_data["Age"].max()  # = 45
#Create Age Bins:
AgeBins = [0, 9, 14, 19, 24, 29, 34, 39, maxAge]
#Create Bin Labels:
AgeLabels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


#Create a list of unique players:
Binned_AgeDemo = UserTable[["SN", "Age"]]
# Binned_AgeDemo

#Bin the age groups:
Binned_AgeDemo["AgeGroup"] = pd.cut(Binned_AgeDemo["Age"], AgeBins, labels = AgeLabels)

# Validate work: 
Binned_AgeDemo

#Total Count of Age Groups:
AgeGroup_df = Binned_AgeDemo.groupby("AgeGroup")
AgeGroup_ttlcount_df = AgeGroup_df.count()
# AgeGroup_ttlcount_df

#Percentage of Age Groups:
#ttl_plyrs #the total number of players
AgeGroup_pcnt_df = AgeGroup_ttlcount_df
# AgeGroup_pcnt_df
AgeGroup_pcnt_df["Percent of Players"] = AgeGroup_pcnt_df["SN"]/ttl_plyrs *100
# AgeGroup_pcnt_df


#########################################################################
#Build a summary table:
#########################################################################
#Summary Table:
SummaryAgeDemo = pd.DataFrame({
    "Total Count" : AgeGroup_ttlcount_df["SN"],
    "Percent of Players": AgeGroup_pcnt_df["Percent of Players"].map("{:.2f}".format)
})

SummaryAgeDemo

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#########################################################################
#Purchasing Analysis (Age)

#Prep the Bin data frame:
Binned_PurchByAge = purchase_data[["Purchase ID", "SN", "Age", "Item ID", "Price"]]
# Binned_PurchByAge.head()

#Bin the age groups:
Binned_PurchByAge["AgeGroup"] = pd.cut(Binned_PurchByAge["Age"], AgeBins, labels = AgeLabels)
# Validate work: 
# Binned_PurchByAge.head()

#Purch count by age groups:
AgeGrp_PurchCount = Binned_PurchByAge.groupby(["AgeGroup"]).count()["Purchase ID"]
AgeGrp_PurchCount

#Total Purchase Value by age groups:
AgeGrp_PurchTotals = Binned_PurchByAge.groupby(["AgeGroup"]).sum()["Price"]
AgeGrp_PurchTotals

#Average Purchase Price by age groups:
AgeGrp_AvgPurch = (AgeGrp_PurchTotals/AgeGrp_PurchCount).map("${:.2f}".format)
AgeGrp_AvgPurch

#########################################################################
#Average Total Purchase per Person:
#1a. use the total purchase value by age groups:
# AgeGrp_PurchTotals_df

#1b. get the count of users within age group:
# From Previous Cells: UserTable = purchase_data.drop_duplicates(subset = "SN", keep = "first")
Binned_SNbyAgeGrp = UserTable[["SN", "Age"]]
# Binned_SNbyAgeGrp
Binned_SNbyAgeGrp["AgeGroup"] = pd.cut(Binned_SNbyAgeGrp["Age"], AgeBins, labels = AgeLabels)
# Binned_SNbyAgeGrp.head()

#1c: User count by age groups:
UserCount_df = Binned_SNbyAgeGrp.groupby(["AgeGroup"]).count()["SN"]
UserCount_df

#2: get the average per user's age group:
Avg_AgeGrp_Per_Person = UserCount_df

Avg_AgeGrp_Per_Person = (AgeGrp_PurchTotals/UserCount_df).map("${:.2f}".format)
Avg_AgeGrp_Per_Person


#########################################################################
#Build a summary table:
#########################################################################
#Summary Table:
SummaryPurchAnalysis = pd.DataFrame({
    "Purchase Count" : AgeGrp_PurchCount,
    "Average Purchase Price": AgeGrp_AvgPurch,  
    "Total Purchase Value": AgeGrp_PurchTotals.map("${:,.2f}".format),
    "Avg Total Purchase per Person" : Avg_AgeGrp_Per_Person
})

SummaryPurchAnalysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#########################################################################
#1a: Get the total number of purchases per person
grp_sn_age_pr = purchase_data[["SN", "Price", "Purchase ID"]]

grp_sn = grp_sn_age_pr.groupby(["SN"])
purch_count_per_sn = grp_sn.count()["Purchase ID"]  #to create a series based on SN
purch_count_per_sn 

#1b: Get the total value of purchases per person
purch_total_per_sn = grp_sn.sum()["Price"]
purch_total_per_sn 

#1c: Get the average total purchase per person:
average_total_purchases = pd.DataFrame({
    "Purchase Count": purch_count_per_sn,
    "Average Purchase Price": purch_total_per_sn/purch_count_per_sn,
    "Total Purchase Value": purch_total_per_sn
    
})    
average_total_purchases

#2a: Sort top down:
sorted_average_total_purchases = average_total_purchases.sort_values(["Total Purchase Value"], ascending = False)
sorted_average_total_purchases.head()

#2b: Format the numbers:
sorted_average_total_purchases["Average Purchase Price"] = sorted_average_total_purchases["Average Purchase Price"].map("${:.2f}".format)
sorted_average_total_purchases["Total Purchase Value"] = sorted_average_total_purchases["Total Purchase Value"].map("${:.2f}".format)

#2c: Display the top 5:
sorted_average_total_purchases.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#########################################################################
#Most Popular Items:
#1: Retrieve the Item ID, Item Name, and Item Price columns:
ItemsPurched_raw = purchase_data[["Item ID", "Item Name", "Price"]]

#2: Group by Item ID and Item Name:
ItemsPurched = ItemsPurched_raw.groupby(["Item ID", "Item Name"])

#3: Perform calculations to obtain purchase count, item price, and total purchase value
#3a:  Purchase Count:
ItemsPurched = ItemsPurched.count()
ItemsPurchedCount = ItemsPurched.rename(columns = {
    "Price": "Purchase Count"
})
ItemsPurchedCount

#3b:  Item Price:
MergedRaw_ItemsPurchedCount = pd.merge(ItemsPurchedCount, ItemsPurched_raw, on = "Item ID")
MergedRaw_ItemsPurchedCount
Merged_ItemsPurchedCount = MergedRaw_ItemsPurchedCount.rename(columns = {
    "Price": "Item Price"
})
Merged_ItemsPurchedCount.head()
Merged_ItemsPurchedCount = Merged_ItemsPurchedCount.drop_duplicates(subset = "Item ID", keep = "first")
Merged_ItemsPurchedCount.head()

#3c:  Total Purchase Value:
Merged_ItemsPurchedCount["Total Purchase Value"] = Merged_ItemsPurchedCount["Purchase Count"] * Merged_ItemsPurchedCount["Item Price"]
Merged_ItemsPurchedCount.head()


#4: sort Purchase Count in descending order:
sorted_ItemsPurched = Merged_ItemsPurchedCount.sort_values(["Purchase Count"], ascending = False)
sorted_ItemsPurched.head()

#5: format the values:
sorted_ItemsPurched["Item Price"] = sorted_ItemsPurched["Item Price"].map("${:.2f}".format)
sorted_ItemsPurched["Total Purchase Value"] = sorted_ItemsPurched["Total Purchase Value"].map("${:.2f}".format)

#6: Reindex to make it look like the example
reIndexed_ItemsPurched = sorted_ItemsPurched.set_index(["Item ID", "Item Name"])
reIndexed_ItemsPurched.head()
```
